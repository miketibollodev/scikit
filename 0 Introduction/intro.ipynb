# Introduction
**Machine Learning:** machine learning deals with building predictive models. One of the goals of machine learning is *generalizing*, which is concluding on new instances.

The challenge of this is that there are many sources of variability. For example, if we are trying to predict if a person is wealthy, their attributes are widely variate like education background, race, marital status, etc.

The typical workflow in machine learning is to use a training dataset to build a model, and apply it to new data to make predictions (with a testing dataset).

**Machine Learning Vocabulary:**
- We use a data matrix, which is a tabular form of data where rows are different observations (*samples*) and columns are different descriptors (*features*).
- In supervised machine learning, we say that data is *annotated* or *labeled*. The purpose is to predict the target/label, or in other words, given a data matrix *X* with *n* observations we predict *y*.
- In unsupervised learning, we are given a data matrix *X* with *n* observations. The goal is to extract some structure from *X* as there are no labels.

## Supervised Learning
In supervised learning we want to predict *y*, which can be discrete or continuous.
- **Classification:** *y* is discrete, made of different classes (i.e., predicting animal species).
- **Regression:** *y* is continuous, which is a numerical quantity (i.e., predicting income).
